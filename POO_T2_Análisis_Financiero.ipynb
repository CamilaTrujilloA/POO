3.- Para anualizar el rendimiento de un per√≠odo, debes capitalizar el rendimiento tantas veces como per√≠odos haya en un a√±o.

Anualiza un 1% mensual, un 4% trimestral y un 0,01% de rentabilidad diaria

```python
def anualizar_rendimiento(rend_periodo, periodos_por_a√±o):
    return (1 + rend_periodo)**periodos_por_a√±o - 1
```

```python
rend_mensual = 0.01
periodos_por_mes = 12
rend_anual_mensual = anualizar_rendimiento(rend_mensual, periodos_por_mes)
print(f'Rendimiento anualizado mensual: {rend_anual_mensual:.4f}')
```

```python
rend_trimestral = 0.04
periodos_por_trimestre = 4
rend_anual_trimestral = anualizar_rendimiento(rend_trimestral, periodos_por_trimestre)
print(f'Rendimiento anualizado trimestral: {rend_anual_trimestral:.4f}') # 4f: son los decimales
```

```python
rend_diario = 0.0001  # 0.01% expresado como fracci√≥n
periodos_por_dia = 252  # Asumiendo los que cotizan las acciones
rend_anual_diario = anualizar_rendimiento(rend_diario, periodos_por_dia)
print(f'Rendimiento anualizado diario: {rend_anual_diario:.6f}')
```

```python
import math
```

```python
def anualizar_volatilidad(vol_periodo, periodos_por_anio):
    return vol_periodo * math.sqrt(periodos_por_anio)
```

```python
from datetime import datetime

import yfinance as yf

class Asset:
    def __init__(self, ticker, start_date,
                 end_date = datetime.today().strftime('%Y-%m-%d')):
        self.ticker = ticker
        self.start_date = start_date
        self.end_date = end_date

    def get_data(self):
        df = yf.download(self.ticker, start = self.start_date, end = self.end_date)
        return df

    def  simple_returns(self, adj = False, cum = False):
        df = self.get_data()
        if adj:
            simple_returns = df["Adj Close"].pct_change().dropna()
        else:
            simple_returns = df["Close"].pct_change().dropna()
        if cum:
            simple_returns = (1 + simple_returns).cumprod() - 1
        return simple_returns

    def  rendimiento_compuesto (rendimientos_simples):
        return (rendimientos + 1).prod() - 1

    def anualizar_rendimiento(rend_periodo, periodos_por_a√±o):
        return (1 + rend_periodo)**periodos_por_a√±o - 1
```

8.- Crea dos activos artificiales que llamar√°s ‚ÄúAA‚Äù y ‚ÄúBB‚Äù, respectivamente. Supongamos que un
cierto activo ‚ÄúAA‚Äù pierde 1% por mes durante 12 meses y un cierto activo ‚ÄúBB‚Äù gana 1% por mes
para los 12 meses. ¬øCu√°l es el activo m√°s vol√°til? Recuerda que la volatilidad se mide a
trav√©s de la desviaci√≥n est√°ndar.

```python
import numpy as pd

#Defino el numero de meses y las tasas tanto de perdida como de ganancia
meses = 12
perdida = 0.01
ganancia = 0.01

#Arreglo para almacenar los valores de del activo AA y del activo BB
val_AA = np.zeros(meses)
val_BB = np.zeros(meses)
```

```python
for mes in range(meses):
    if mes == 0:
        val_AA[mes] = 100
        val_BB[mes] = 100
    else:
        val_AA[mes] = val_AA[mes - 1] * (1 - perdida)
        val_BB[mes] = val_BB[mes - 1] * (1 + ganancia)
```

```python
#Para calcular la desviaci√≥n est√°ndar
ds_AA = np.std(val_AA)
ds_BB = np.std(val_BB)
```

```python
#Imprimimos los valores de los activos
print("Valores finales de AA:", val_AA)
print("Valores finales de BB:", val_BB)
```

```python
#Imprimimos la volatibilidad de los activos
print("Desviaci√≥n est√°ndar de AA:", ds_AA)
print("Desviaci√≥n est√°ndar de BB:", ds_BB)
```

El activo BB tuvo una mayor volatilidad ya que su desviaci√≥n est√°ndar es mayor.
