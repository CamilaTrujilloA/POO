3.- Anualizamos la volatilidad escalándola (multiplicándola) por la raíz cuadrada del número de períodos por observación.

```python
import pandas as pd
import random
import numpy as np

# Fijar los valores aleatorios
random.seed(12)

# Generar valores aleatorios
valores_aleatorios = [random.uniform(8, 13) for _ in range(12)]

# Crear un diccionario con las etiquetas y los valores
data = {'Precios': valores_aleatorios}

# Crear el DataFrame
precios_muestra = pd.DataFrame(data)
precios_muestra.index = [f"Mes {i+1}" for i in range(12)]

df=precios_muestra
df
```

```python
#sacamos los rendimientos del data frame
rendimientos=df.pct_change().dropna()
rendimientos
```

```python
#sacamos el promedio de los renimientos
media=rendimientos.mean()
media
```

```python
#sacamos las deviaciones estandar del data frame
desviaciones=(rendimientos-media*(rendimientos))**2
desviaciones
```

```python
#sacamos la media de las desviaciones estandar
desv_mean=desviaciones.mean()
desv_mean
```

```python
#sacamos la volatilidad del data frame
volatilidad=desv_mean**1/2
volatilidad
```

```python
#Anualizar la volatilidad
anualizar_vol=volatilidad*(12**1/2)
anualizar_vol
```
