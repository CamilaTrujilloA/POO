```python
# 10.- Explica el siguiente código:

df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'], 'key2' : ['one', 'two', 'one', 'two', 'one'], 'data1' : np.random.randn(5),'data2' : np.random.randn(5)})
df
```

En el codigo anterior:

pd.DataFrame: se refiere a la estructura de datos proporcionada por la biblioteca pandas.

'key1' : ['a', 'a', 'b', 'b', 'a']: Da el nombre de la primer columna y los valores que la integran.

'key2' : ['one', 'two', 'one', 'two', 'one']: Da el nombre de la segunda columna y los valores que la integran.

data1' : np.random.randn(5): Da el nombre de la tercera columna y los valores que la integran salen de una función que genera un arreglo de tamaño 5 con valores aleatorios.

'data2' : np.random.randn(5): Da el nombre de la cuarta columna y los valores que la integran salen de una función que genera un arreglo de tamaño 5 con valores aleatorios.

```python
# (a) Quita la columna de "key 2".

df_a = df.drop(columns=["key2"])
df_a
```
