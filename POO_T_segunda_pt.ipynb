1. Crea definiciones que calculen el área de un círculo y el área de un rectángulo, respectivamente.



```python
# importamos el modulo math.
# generamos la definición para calcular el area de un circulo.
import math

def Circulo(r):
  C =math.pi*(r**2)
  return C
```

```python
# comprobación de la definición.
Circulo(2)
```

```python
# generamos la definición para calcular el area de un rectángulo.
def Rectangulo(b,h):
  R =b*h
  return R
```

```python
# comprobación de la definición.
Rectangulo(2,3)
```

2. Crea una definición que calcule la raíz cuadrada de un número.

```python
# generamos la definición para calcular la raiz cuadrada de un número.
def Raiz(a):
  raiz= a**(1/2)
  return raiz
```

```python
# comprobación de la definición.
Raiz(9)
```
