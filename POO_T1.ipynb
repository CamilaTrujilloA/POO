Tarea: subir una base de datos de 10x10

```python
import random

# Lista de nombres de Tenistas
nombres = ["Alcaraz", "Djokovic", "Medvedev", "Tsitsipas", "Rune", "Sinner", "Ruud", "Rublev", "Fritz", "Tiafoe"]

# Generar matriz de datos aleatorios en este caso predicciones de juegos ganados en los proximos 10 años
matriz_datos = []
for _ in range(10):
    filas= [random.randint(10, 70) for _ in range(10)]
    matriz_datos.append(filas)
```

```python
import pandas as pd
matriz = pd.DataFrame(matriz_datos, columns=[f"AÑO {i+1}" for i in range(10)], index=nombres)
matriz
```
