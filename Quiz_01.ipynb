```python
# 1.- Crea un diccionario que contenga 3 llaves que sean de tipo string y cuyos valores sean una lista que contenga 5 valores (3 numéricos y 2 flotantes).

Mascotas=({"Snoopy":[2,3,4,6.7,8.5], "Bolt":[3,4,7,8.9,10.5],"Pongo":[5,6,4,15.3,9.8]})
```

```python
type(Mascotas)
```

```python
# 2.- A partir del diccionario anterior, muestra a sus ítems. ¿Qué tipo de objeto se obtiene?

Mascotas.items()
```

```python
type(Mascotas.items())
```

```python
# 3.- Crea un panel de datos usando tu diccionario. Realiza tres segregaciones distintas.
import pandas as pd

Info_Mascotas=pd.DataFrame(Mascotas, index=["Familiares","Premios al dia","Edad","Peso","Velocidad en km/h"])
Info_Mascotas
```

```python
# tiras columas, dejas de usar unas
sin_bolt=Info_Mascotas.drop(columns=["Bolt"])
sin_bolt
```

```python
# selección por indices (renglones)
sin_fam_ni_prem=Info_Mascotas.iloc[2:6]
sin_fam_ni_prem
```

```python
# es para cortar y seleccionar columnas
Info_Mascotas.iloc[:,[1,2]]
```

```python
# 4.- Realiza un breve análisis descriptivo de tu panel de datos. Explica los valores que obtienes.
# estadistica descriptiva
Info_Mascotas.describe()
```

En la tabla del analisis descriptivo podemos ver que como "count" nos dce cauntos datos tiene cada string, mean nos dice el promedio de los datos, nos da el calculo de la desviación estandar, tambien nos da los cuartiles es decir que valor se encuentra en el 25%, 50% y 75% por ultimo el vamor maximo de los datos ingresados.

```python
# 5.- Realiza dos filtros de búsqueda sobre tu panel de datos.

# Solo salen los valores mayores a 7 para Pongo
Info_Mascotas[Info_Mascotas['Pongo']>7]
```

```python
# el eje de las renglones es el eje 1 (axis=1)
# mean es para sacar promedios
Info_Mascotas.mean(axis=1)
```
